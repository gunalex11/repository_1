```python
import math
import cmath
import numpy as np
import matplotlib.pyplot as plt

iterations = 20

def my_exp(x):
    """Вычисление экспоненты при помощи частичного суммирования ряда Тейлора"""
    x_pow = x
    multiplier = 1
    partial_sum = 1+x
    for n in range(1, iterations):
        x_pow *= x
        multiplier *= 1 / (n+1)
        partial_sum += x_pow * multiplier
    
    return partial_sum

print("Библиотечная экспонента:      ", math.e**2.5)
print("Экспонента через ряд тейлора: ", my_exp(2.5))

%matplotlib inline

plt.figure(1) # графики библиотечной и моей экспоненты приблизительно совпадают на промежутке от 0 до 5
arguments1 = np.r_[0:5:0.01]
plt.plot(arguments1, [math.e**a for a in arguments1])
plt.plot(arguments1, [my_exp(a) for a in arguments1])

plt.figure(2) #  на промежутке от 0 до 20 уже видно, что графики библиотечной и моей экспоненты расходятся
arguments2 = np.r_[0:20:0.01]
plt.plot(arguments2, [math.e**a for a in arguments2])
plt.plot(arguments2, [my_exp(a) for a in arguments2])

plt.show()
```
