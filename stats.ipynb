```python
%matplotlib inline  

import pandas
import matplotlib
import numpy
import matplotlib.pyplot as plt

# Данные из лабораторной работы "Моделирование недослучайной величины"
experiments = [7.961, 7.97, 7.975, 7.982, 7.984,
               7.99, 7.993, 7.994, 7.995, 7.995,
               7.996, 7.997, 7.997, 8, 8.002,
               8.005, 8.008, 8.009, 8.012, 8.013,
               8.013, 8.013, 8.013, 8.014, 8.016,
               8.016, 8.016, 8.019, 8.02, 8.02,
               8.024, 8.024, 8.025, 8.026, 8.027,
               8.027, 8.028, 8.028, 8.03, 8.031,
               8.031, 8.031, 8.032, 8.033, 8.034,
               8.034, 8.037, 8.04, 8.041, 8.052]

plt.plot(experiments)
plt.title('Моделирование недослучайной величины')
plt.show()
```

#### Не стоит пугаться! Просто данные отсортированы по возрастанию.
#### На данном этапе уже можно предсказать более-менее хорошей формы пик у функции распределения
#### (в низких и в высоких значениях график резче, чем у среднего значения).

## Поглядим на распределение

```python
df = pandas.DataFrame(data={'experiments': experiments})

df.plot.kde()
```

### Бингооо! Похоже на нормальное распределение!

### Теперь оценим "Гауссовость" наших данных количественно

```python
from scipy import stats

dfe = df['experiments']

print("К-С-статистика:",stats.kstest(dfe, 'norm', (dfe.mean(), dfe.std()))[0])
print("P-значение:    ",stats.kstest(dfe, 'norm', (dfe.mean(), dfe.std()))[1])
```

#### Имеем довольно низкую К-С-статистику, и, вместе с тем, довольно высокое P-значение. Это позволяет с помощью критерия Колмогорова-Смирнова утверждать, что наши данные распределяются более-менее нормально.
